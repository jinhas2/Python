### 시각화를 위한 Matplotlib 패키지 활용하기

```python
#mat plot lib 으로 끊어서 읽는다. 
import matplotlib.pyplot as plt
```

Style 적용하기

matplotlib style 검색 후 원하는 스타일 이름 정하기 


```python
plt.style.use('classic')
```

```python
import numpy as np 
```

```python
x = np.linspace(0,10,100) # 0에서 10까지 100개의 데이터를 생성. 
print(x)
```

```python
plt.plot(x, np.sin(x))
```

```python
plt.plot(x, np.cos(x))
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig = plt.figure() #그림판 만들기 
plt.plot(x, np.sin(x), '-') 
plt.plot(x, np.cos(x), '--') 
fig.savefig('my_figure.png')
```

```python
fig.canvas.get_supported_filetypes() 
```

```python
plt.plot(x, np.sin(x), "-", linewidth = 1, label = "sin(x)")
plt.plot(x, np.cos(x), "--", linewidth = 1, label = "cos(x)")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.title("Sine & Cosine Curve")
plt.xlabel('x')
plt.legend() 
```

```python
plt.subplot(2,1,1) #2줄 컬럼 1개 짜리 
plt.plot(x, np.sin(x))
plt.subplot(2,1,2)
plt.plot(x, np.cos(x)) 
```

```python
plt.subplot(2,2,1) 
plt.plot(x, np.sin(x))
plt.subplot(2,2,2)
plt.plot(x, np.cos(x)) 
plt.subplot(2,2,3) 
plt.plot(x, np.sin(x))
plt.subplot(2,2,4)
plt.plot(x, np.cos(x)) 
```

축을 설정해서 그림 그리기 

```python
plt.style.use('seaborn-whitegrid') 
```

```python
fig = plt.figure() 
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x, np.sin(x)) 
```

다양한 선 스타일 

```python
plt.plot(x, np.sin(x-0), color='blue')                # CMYK , RGB 
plt.plot(x, np.sin(x-1), color='g')                   # 짧은 색상 코드, 약자로 표현가능, black은 k로 사용 
plt.plot(x, np.sin(x-2), color='0.75')                # 0-1사이의 회색조를 표현한다. 
plt.plot(x, np.sin(x-3), color='#FFDD44')             # 16진수로 표현한 RGB
plt.plot(x, np.sin(x-4), color=(1.0,0.2,0.3))         # RGB값을 튜플로 표현, 10진수로 표현 
plt.plot(x, np.sin(x-5), color='chartreuse')          # 원하는 이름을 선택
```

```python
plt.plot(x, x+0, linestyle='solid')                # 실선 
plt.plot(x, x+1, linestyle='dashed')               # 점선 
plt.plot(x, x+2, linestyle='dashdot')
plt.plot(x, x+3, linestyle='dotted')

plt.plot(x, x+4, linestyle='-')
plt.plot(x, x+5, linestyle='--')
plt.plot(x, x+6, linestyle='-.')
plt.plot(x, x+7, linestyle=':')
```

```python
plt.plot(x, x+0, '-g')                            # solid green 
plt.plot(x, x+1, '--c')                           # dashed cyan 
plt.plot(x, x+2, '-.k')                           # dashdot black 
plt.plot(x, x+3, ':r')                  
```

플롯 조정하기 : 축 경계

```python
plt.plot(x, np.sin(x)) 
plt.xlim(-1,11)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(x, np.sin(x)) 
plt.xlim(10,0) #좌우반전 
plt.ylim(1.5,-1.5) #상하반전 
```

```python
plt.plot(x, np.sin(x))
plt.axis([-1,11,-1.5,1.5]) 
```

```python
plt.plot(x, np.sin(x))
plt.axis('tight') # 전체 그래프에 타이트하게 꽉 채워서 출력. 
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal') # x축과 y축의 단위를 독같이 만든다. 
```

플롯에 레이블 붙이기

```python
plt.plot(x, np.sin(x))
plt.title('A Sine Curve')
plt.xlabel('x')
plt.ylabel('y = Sin(x)')
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)') #선의 이름 지정 
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend() 
```

산점도의 출력 

```python
x = np.linspace(1,10,30)  #30개
y = np.sin(x)

plt.plot(x, y, 'o', color='black') # 'o', ',', 'x', '+', 'v', '<', '>','s','d' 다양한 마커가 존재함. 
```

```python
rng = np.random.RandomState(0)

for marker in ['o',',','x','+','v','<','>','s','d']:
  plt.plot(rng.rand(5), rng.rand(5), marker, label='marker={0}'.format(marker))
  plt.legend() 
  plt.xlim(0,1.8)
```

```python
plt.plot(x,y,'-ok')
```

```python
plt.plot(x,y,'-p', color='gray', markersize=15, linewidth=4, 
         markerfacecolor='white', 
         markeredgecolor='red', 
         markeredgewidth='2') #p는 오각형. 
```

```python
plt.scatter(x,y, marker='o', s=50)
```
