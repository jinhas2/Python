```python
import requests
import bs4 
```

```python
url = 'https://search.shopping.naver.com/search/category/100005307'
result = requests.get(url).text 
result
```

```python
bsObj = bs4.BeautifulSoup(result, 'html.parser') #html해석기, result안에 내용을 해석하겠다. 
```

```python
bsObj.find('a', {'class':'basicList_link__JLQJf'}) #a태그에서 내용을 찾아라, 개발자도구에서 찾은 값을 더블클릭하면 복사가능.
```

```python
bsObj.find('a', {'class':'basicList_link__JLQJf'}).text
```

```python
items = bsObj.find_all('a', {'class':'basicList_link__JLQJf'})
```

```python
for item in items:
  print(item.text)
```

```python
print(items[0].text)
print(items[1].text)
print(items[2].text)
print(items[3].text)
```

```python
for i in items[:3]:
  print(i.text)
```
