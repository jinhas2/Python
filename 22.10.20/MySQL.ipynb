```python
import pymysql
```

```python
host_name = 'labuser5mysql.mysql.database.azure.com'
username = 'jinhas2'
password = ' ' #password는 보안 
database_name = 'classicmodels' 
```

```python
from IPython.lib import passwd
# 방화벽 규칙 적용. 
db = pymysql.connect(
    host=host_name,
    user=username,
    port=3306, #특별히 지정안하면 mysql은 3306
    passwd=password,
    db=database_name,
    charset='utf8' ,
    ssl={"fake_flag_to_enable_tls":True}  
)
```

```python
import pandas as pd
```

```python
SQL = 'SELECT * FROM employees'
df = pd.read_sql(SQL, db) 
df 
```

```python
# TINYINT는 8비트 
SQL = '''
CREATE TABLE tempTable(
  id TINYINT NOT NULL AUTO_INCREMENT, 
  user_name varchar(10) NOT NULL,
  phone varchar(30) NULL ,
  PRIMARY KEY(id) 
) 
''' 
cursor = db.cursor()
cursor.execute(SQL) 
cursor.close()
```
